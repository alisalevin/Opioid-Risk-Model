```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
model = pd.read_csv("model_test.csv")
```

```python
# Separate Data into 5 States
ohio = model.loc[model['State'] == 'OH']
virginia = model.loc[model['State'] == 'VA']
kentucky = model.loc[model['State'] == 'KY']
pennsylvania = model.loc[model['State'] == 'PA']
west_virginia = model.loc[model['State'] == 'WV']
```

```python
# Ohio Prediction Plot
ohio.plot(kind="scatter", x="Longitude", y="Latitude", alpha=0.4,
    s=ohio["Model Output"]*1000, label="Model Output", figsize=(10,7),
    c="Model Output", cmap=plt.get_cmap("jet"), colorbar=True,
    sharex=False)
# Marker for target at-risk county that we used in the formulation of the model: Cuyahoga and Hamilton
plt.annotate(xy=[-81.9,41.33], s="O")
plt.annotate(xy=[-84.75,39.1], s="O")

# Marker for next-worst test counties we wanted to predict as most likely to have an opioid crisis (Franklin and Montgomery)
plt.annotate(xy=[-83.2,40.05], s="X")
plt.annotate(xy=[-84.45,39.8], s="X")
plt.legend()
```
